```python
def postorder(t):
    nodes = []
    _postorder(t, nodes)
    return nodes
def _postorder(t, nodes):
    if t is None:
        return
    for sub in t.kids:
        _postorder(sub, nodes)
    nodes.append(t)
```

```python
def leaves(t):
    nodes = []
    _leaves(t, nodes)
    return nodes
def _leaves(t, nodes):
    if t is None:
        return
    if len(t.kids)==0:
        nodes.append(t)
        return
    for sub in t.kids:
        _leaves(sub, nodes)
```

```python
nodes = postorder(root)
atoms = leaves(root)
atoms
```

```python
# map tokens to nodes
tok2node = {}
for nd in atoms:
    tok2node[nd.token] = nd
```

```python
def walk(t, pre=lambda x:None, post=lambda x:None):
    if t is None:
        return
    pre(t)
    for sub in t.kids:
        walk(sub, pre, post)
    post(t)
```

```python
import matplotlib
import matplotlib.patches as patches
from matplotlib import pyplot as plt
def textdim(s, fontsize=11):
    fig, ax = plt.subplots(1,1)
    t = ax.text(0, 0, s, bbox={'lw':0}, fontsize=fontsize)
    plt.savefig("/tmp/junk")
    plt.close()
    bb = t.get_bbox_patch()
    w, h = bb.get_width(), bb.get_height()
    return w, h
```
